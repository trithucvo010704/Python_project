```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. Basic Matplotlib visualization:

#### 1.1. Bar plot:

```python
x = np.array(['Q1', 'Q2', 'Q3','Q4'])
y = np.array([ 234.0, 254.7, 144.6, 317.6])
plt.bar(x,y,color= 'purple')
plt.title('BAR CHART')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

#### 1.2. Histogram:

```python
x = np.random.randn(10000)
plt.hist(x,bins=20, color='red',density=True)
plt.title('Histogram')
plt.show()
```

```python
data = np.random.randn(10000)
plt.hist(data, bins=100, color='blue',density=False,alpha=0.3)
plt.title('Histogram')
plt.show()
```

#### 1.3. Box plot:
##### Box plot (hay box-and-whisker plot) là một dạng biểu đồ được sử dụng để hiển thị phân bố của một tập dữ liệu thông qua các quartile. Nó cung cấp cái nhìn tổng quan về tập dữ liệu, bao gồm cả những điểm ngoại lai.

```python
x = np.random.randn(10000) * 10 + 3 
y = np.random.randn(10000) * 10 + 5 
z = np.random.randn(10000) * 10 + 1
plt.boxplot([x,y,z] , 0)
plt.title('BOX PLOT')
plt.show() 
```

```python
plt.boxplot([x,y,z],0,vert=False)
plt.title('BOX PLOT')
plt.show()
```

#### 1.4. Line plot:

```python
x= np.linspace(0,10,100) 
y  =np.sin(x)
plt.plot(x,y,color = 'orange',linewidth=2 , linestyle ='--')
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='purple',linestyle='-',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.xlim([0,5])                                     # Horizontal axis limits. ( dat gioi han truc x)
plt.ylim([-2,+2])                                   # Vertical axis limits. ( dat gioi han truc y )
plt.show() 
```

#### 1.5. Scatter plot 1:

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x, y, color='blue', marker='s', linestyle='none', markersize=10, alpha=1)
# plt.plot(x,y,color='red',marker='o',linestyle='none',markersize=5, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.scatter(x,y,c='red',marker='.',alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.scatter(x,y,c='orange',marker='s',alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```
